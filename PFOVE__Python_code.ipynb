```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Mean response function (m) and standard deviation function (s)
def mean_response(x):
    x1, x2 = x
    return (3 * x1 + 3 * x2 + 6) ** 2

def std_dev_response(x):
    x1, x2 = x
    return ((x1 + x2) ** 2 + 2) ** 2

def objective(x, p1, p2):
    return (abs(mean_response(x)) / 100) ** p1 + (abs(std_dev_response(x)) / 100) ** p2

# Optimization settings
bounds = [(-1, 1), (-1, 1)]
initial_x = [0, 0]

# List to store Pareto solutions
pareto_solutions = []

# Try various combinations of p1 and p2
p1_values = np.arange(0.05, 3, 0.1)
p2_values = np.arange(0.05, 3, 0.1)

for p1 in p1_values:
    for p2 in p2_values:
        result = minimize(objective, initial_x, args=(p1, p2), bounds=bounds, method='SLSQP')
        if result.success:
            m_opt = mean_response(result.x)
            s_opt = std_dev_response(result.x)
            x_opt = result.x
            pareto_solutions.append((m_opt, s_opt, x_opt))

# Find the solution with the smallest sum of m + s
min_sum_solution = min(pareto_solutions, key=lambda tup: tup[0] + tup[1])
m_min, s_min, x_min = min_sum_solution

# Visualize the results
if pareto_solutions:
    m_values, s_values, x_values = zip(*pareto_solutions)
    plt.scatter(m_values, s_values, s=5, color='b', alpha=0.5, label='Pareto Solutions')
    plt.plot(m_min, s_min, 'ro', label=r'min$(\hat{f}_1(\mathbf{x}^*) + \hat{f}_2(\mathbf{x}^*))$', markersize=8)
    plt.text(m_min + 0.5, s_min + 0.5, r"$min (\hat{f}_1(\mathbf{x}^*) + \hat{f}_2(\mathbf{x}^*) )$", fontsize=10, color='red')
    plt.ylabel(r"$\hat{f}_2(\mathbf{x}^*)$")
    plt.xlabel(r"$\hat{f}_1(\mathbf{x}^*)$")
    plt.xlim(0, 30)
    plt.ylim(0, 30)
    plt.legend()
    plt.show()

# Print all optimized solutions
for i, (m, s, x) in enumerate(pareto_solutions):
    print(f"Solution {i+1}: m = {m:.2f}, s = {s:.2f}, x = {x}")

print("\n✅ Solution with the smallest value of m + s:")
print(f"m = {m_min:.2f}, s = {s_min:.2f}, m + s = {m_min + s_min:.2f}, x = {x_min}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Mean response function (m) and standard deviation response function (s)
def mean_response(x):
    x1, x2, x3 = x
    return (0.5 * x1**2 - 0.5 * x2**2 + 0.5 * x3**2 + x1 * x3 - 4) ** 2

def std_dev_response(x):
    x1, x2, x3 = x
    return (x1**2 - x2**2 + 3 * x3) ** 2

# Objective function: normalized and weighted combination
def objective(x, p1, p2):
    return (abs(mean_response(x)) / 100) ** p1 + (abs(std_dev_response(x)) / 100) ** p2

# Optimization settings
bounds = [(-1, 1), (-1, 1), (-1, 1)]
initial_x = [1,1,1]

# List to store all solutions
all_solutions = []

# Try various (p1, p2) combinations
p1_values = np.arange(0.05, 3, 0.05)
p2_values = np.arange(0.05, 3, 0.05)

for p1 in p1_values:
    for p2 in p2_values:
        result = minimize(objective, initial_x, args=(p1, p2), bounds=bounds, method='SLSQP')
        if result.success:
            m_opt = abs(mean_response(result.x))
            s_opt = abs(std_dev_response(result.x))
            x_opt = result.x
            all_solutions.append((m_opt, s_opt, x_opt))

# Pareto front filtering
def is_dominated(m, s, others):
    for m2, s2 in others:
        if (m2 <= m and s2 <= s) and (m2 < m or s2 < s):
            return True
    return False

def filter_pareto_front(m_values, s_values):
    pareto_m = []
    pareto_s = []
    pareto_indices = []
    points = list(zip(m_values, s_values))

    for i, (m, s) in enumerate(points):
        others = points[:i] + points[i+1:]
        if not is_dominated(m, s, others):
            pareto_m.append(m)
            pareto_s.append(s)
            pareto_indices.append(i)

    return pareto_m, pareto_s, pareto_indices

# Extract m, s, x from solutions
if all_solutions:
    m_values, s_values, x_values = zip(*all_solutions)

    # Find Pareto-optimal solutions only
    pareto_m, pareto_s, pareto_indices = filter_pareto_front(m_values, s_values)

    # Compute total risks (m + s) for Pareto points only
    total_risks = [m + s for m, s in zip(pareto_m, pareto_s)]
    min_index = np.argmin(total_risks)
    min_m = pareto_m[min_index]
    min_s = pareto_s[min_index]
    min_x = x_values[pareto_indices[min_index]]

    # Plot Pareto-optimal solutions
    plt.figure(figsize=(7, 6))
    plt.scatter(pareto_m, pareto_s, s=10, color='blue', alpha=0.7, label='Pareto Solutions')
    plt.scatter(min_m, min_s, color='red', s=60, label=r'$min (\hat{f}_1(\mathbf{x}^*) + \hat{f}_2(\mathbf{x}^*))$', zorder=5)
    plt.text(min_m + 0.5, min_s + 0.5, r"$min (\hat{f}_1(\mathbf{x}^*) + \hat{f}_2(\mathbf{x}^*))$", fontsize=10, color='red')
    plt.ylabel(r"$\hat{f}_2(\mathbf{x}^*)$")
    plt.xlabel(r"$\hat{f}_1(\mathbf{x}^*)$")
    plt.xlim(0, 17)
    plt.ylim(0, 17)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Print result
    print(f"✅ Minimum Total Risk T = f₁ + f₂: {min_m + min_s:.5f}")
    print(f"Corresponding mean (m): {min_m:.5f}")
    print(f"Corresponding std deviation (s): {min_s:.5f}")
    print(f"Corresponding input x: {min_x}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Mean response function (objective 1)
def mean_response(x):
    x1, x2, x3 = x
    y1_mu = 104.867 - 3.148*x1 + 2.379*x1*x2 - 0.35*x1*x3
    y2_mu = 70.4514 - 0.3488*x1 + 3.592*x2 - 0.449*x1*x3 + 0.6144*x2*x3
    term1 = (y1_mu - 103)**2
    term2 = (y1_mu - 103) * (y2_mu - 73)
    term3 = 2.5 * (y2_mu - 73)**2
    return term1 + term2 + term3

# Standard deviation response function (objective 2)
def std_dev_response(x):
    x1, x2, x3 = x
    y1r_sq = (1.613 - 0.254 * x2 + 1.25 * x3)**2
    y2r_sq = 2.5 * (1.7288 + 0.417 * x1 + 1.214 * x3 + 0.2625 * x1 * x3)**2
    cov_2x = (3.7118 + 1.2034 * x1 + 3.5550 * x3 + 1.1611 * x1 * x3)
    return y1r_sq + y2r_sq + cov_2x

# Weighted scalarization objective function
def objective(x, p1, p2):
    return (abs(mean_response(x)) / 188.641)**p1 + (abs(std_dev_response(x)) / 63.5602)**p2

# Function to check if a point is dominated by any other
def is_dominated(m, s, others):
    for m2, s2 in others:
        if (m2 <= m and s2 <= s) and (m2 < m or s2 < s):
            return True
    return False

# Filter non-dominated solutions (Pareto front)
def filter_pareto_front(m_values, s_values):
    pareto_m, pareto_s, pareto_indices = [], [], []
    points = list(zip(m_values, s_values))
    for i, (m, s) in enumerate(points):
        others = points[:i] + points[i+1:]
        if not is_dominated(m, s, others):
            pareto_m.append(m)
            pareto_s.append(s)
            pareto_indices.append(i)
    return pareto_m, pareto_s, pareto_indices

# Optimization bounds and initial guess
bounds = [(-1, 1), (-1, 1), (-1, 1)]
initial_x = [0, 0, 0]
pareto_solutions = []

# Grid search over weights (p1, p2)
p1_values = np.arange(0.05, 3, 0.05)
p2_values = np.arange(0.05, 3, 0.05)

for p1 in p1_values:
    for p2 in p2_values:
        result = minimize(objective, initial_x, args=(p1, p2), bounds=bounds, method='SLSQP')
        if result.success:
            m = mean_response(result.x)
            s = std_dev_response(result.x)
            x = result.x
            pareto_solutions.append((m, s, x))

# Extract and filter Pareto-optimal solutions
m_values, s_values, x_values = zip(*pareto_solutions)
m_values = np.abs(np.array(m_values))
s_values = np.array(s_values)
pareto_m, pareto_s, pareto_indices = filter_pareto_front(m_values, s_values)
filtered_m = [m_values[i] for i in pareto_indices]
filtered_s = [s_values[i] for i in pareto_indices]
filtered_x = [x_values[i] for i in pareto_indices]
filtered_pareto_solutions = list(zip(filtered_m, filtered_s, filtered_x))

# Select representative solutions
min_sum_solution = min(filtered_pareto_solutions, key=lambda tup: tup[0] + tup[1])
m_min, s_min, x_min = min_sum_solution

min_m_solution = min(filtered_pareto_solutions, key=lambda tup: tup[0])
m_only_min, s_at_m_min, x_at_m_min = min_m_solution

min_s_solution = min(filtered_pareto_solutions, key=lambda tup: tup[1])
m_at_s_min, s_only_min, x_at_s_min = min_s_solution

# Find special points with m/s close to given ratios
target_ratios = [1, 2, 0.5]
special_points = []
special_x = []

for r in target_ratios:
    idx = np.argmin(np.abs(np.array(filtered_m) / np.array(filtered_s) - r))
    special_points.append((filtered_m[idx], filtered_s[idx]))
    special_x.append(filtered_x[idx])

# Plotting
plt.figure(figsize=(7, 6))
plt.scatter(filtered_m, filtered_s, s=8, color='blue', alpha=0.6, label='Pareto Solutions')

# Highlight representative solutions
plt.scatter(m_min, s_min, color='green', s=40, label=fr"min ($\hat{{f}}_1(\mathbf{{x}}^*)+\hat{{f}}_2(\mathbf{{x}}^*)$)", zorder=5)
plt.scatter(m_only_min, s_at_m_min, color='red', marker='*', s=40, label=fr"min $\hat{{f}}_1(\mathbf{{x}}^*)$", zorder=5)
plt.scatter(m_at_s_min, s_only_min, color='orange', marker='*', s=40, label=fr"min $\hat{{f}}_2(\mathbf{{x}}^*)$", zorder=5)

# Highlight special ratio points
sp_m, sp_s = zip(*special_points)
plt.scatter(sp_m, sp_s, s=50, color='r', marker='o')
origin = np.array([0, 0])
for point in special_points:
    plt.plot([origin[0], point[0]], [origin[1], point[1]], 'r--', linewidth=1)
    plt.text(point[0] + 0.1, point[1]+0.2, f"({point[0]:.2f}, {point[1]:.2f})", fontsize=7, color='r')

plt.xlabel(r"$\hat{f}_1(\mathbf{x}^*)$")
plt.ylabel(r"$\hat{f}_2(\mathbf{x}^*)$")
plt.legend()
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.tight_layout()
plt.show()

# Print information
print("\n✅ Solution minimizing m + s:")
print(f"m = {m_min:.5f}, s = {s_min:.5f}, m + s = {m_min + s_min:.5f}, x = {x_min}")

print("\n✅ Solution minimizing m only:")
print(f"m = {m_only_min:.5f}, s = {s_at_m_min:.5f}, x = {x_at_m_min}")

print("\n✅ Solution minimizing s only:")
print(f"m = {m_at_s_min:.5f}, s = {s_only_min:.5f}, x = {x_at_s_min}")

# Print information for special ratio points
for ratio, x_val, (m_val, s_val) in zip(target_ratios, special_x, special_points):
    print(f"\n⭐ Ratio ~ {ratio}: m = {m_val:.5f}, s = {s_val:.5f}, x = {x_val}")
```
